# Pyplot

matplotlib.pyplot is a collection of command style fxns.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])       
# if one list of values is given then it is consider as the values of y
# y values and if x values are not given then indices becomes the x values
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title('MyPlot')
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

```python
# Third optional argument which is the format string that indicates the color and line type of the plot

plt.plot([1,2,3,4], [1,4,9,16], 'ro')       # r represent the red color and o represent put dots rather line
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0., 5., 0.2)
# blue dashes, red squares and green triangles
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()   # add legend based on line labels (Make plot really readable)
plt.show()
```

```python
# Controlling line properties

x= [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y, linewidth=5.0)
plt.show()
```

```python
# use the setp()    SetProperties

x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1, y1, x2, y2)

# use keyword arguments
plt.setp(lines[0], color='r', linewidth = 2.0)

# or MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)

plt.grid()
plt.show()
```

```python
# Official documenatation example: https://matplotlib.org/users/pyplot_tutorial.html
# The figure() command here is optional because figure(1) will be created by default, just as a subplot(111) will be 
# created by default if you don’t manually specify any axes. The subplot() command specifies numrows, numcols, fignum 
# where fignum ranges from 1 to numrows*numcols. The commas in the subplot command are optional if numrows*numcols<10.
# So subplot(211) is identical to subplot(2, 1, 1). You can create an arbitrary number of subplots and axes. If you want 
# to place an axes manually, i.e., not on a rectangular grid, use the axes() command, which allows you to specify the 
# location as axes([left, bottom, width, height]) where all values are in fractional (0 to 1) coordinates.

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# plt.figure(1)
# plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

# plt.subplot(212)
# plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.subplot(223)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(224)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```
